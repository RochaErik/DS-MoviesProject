```python
df_best['profit_musd'] = df.revenue_musd.sub(df.budget_musd)
df_best['return'] = df.revenue_musd.div(df.budget_musd)
```

We have movies with high average rating but with a very few number of votes

```python
df_best.sort_values(by='ROI', ascending=False)
```

As we see, the dataset contains movies with high ROI values but without any budget value. (ROI = revenue/budget)

```python
df_best.loc[df_best.Budget >= 5].sort_values(by='ROI', ascending=False)
```

```python
df_best.Budget.fillna(0, inplace=True)
df_best.Votes.fillna(0, inplace=True)
```

```python
df_best.info()
```

```python
def best_worst(n, by, ascending=False, min_budget=0, min_votes=0):
    df2 = df_best.loc[(df_best.Budget >= min_budget) & (df_best.Votes >= min_votes), 
    ['',by]].sort_values(by=by, ascending=ascending).head(n).copy()

    return HTML(df2.to_html(escape=False))
```

**Top 5 movies in revenue**

```python
best_worst(5, 'Revenue')
```

**Top 5 movies in budget**

```python
best_worst(5, 'Budget')                                                                                                 
```

**Worst 5 movies by profit**

```python
best_worst(5, 'Profit', True)
```

**Top 5 movies by average rating**

```python
best_worst(5, 'Average Rating', min_votes=20)
```

**Worst 5 movies by average rating**

```python
best_worst(5, 'Average Rating', ascending=True, min_votes=20)
```

**Most successful actors**

```python
df.cast
```

```python
df
```

```python
#each movie has his own unique id so we can set the index = id

df.set_index('id', inplace=True)
```

```python
df.cast
```

```python
act = df.cast.str.split('|', expand=True)
act
```

```python
act.stack()
```

```python
#convert each columns to index (stack()) then remove the second column (reset_index(level=1, drop=True)) and finally converts all to a dataframe
act = act.stack().reset_index(level=1, drop=True).to_frame()
act
```

```python
act.columns = ['Actors']
```

```python
right = df[['title', 'revenue_musd', 'vote_average', 'popularity']]
right
```

```python
#merging on the index of both tables
act = act.merge(right, how='left', left_index=True, right_index=True)
act.head(20)
```

```python
act.Actors.nunique()
```

```python
act.Actors.unique()
```

```python
act.Actors.value_counts().head(20)
```

```python
plt.figure(figsize=(10,5))
act.Actors.value_counts().head(20).plot(kind='bar', fontsize=15)
plt.title('Actors', fontsize=15)
plt.ylabel('Number of movies', fontsize=15)
plt.show()
```

```python
actors_data = act.groupby('Actors').agg(Total_revenue = ('revenue_musd', 'sum'),
                          Median_revenue = ('revenue_musd', 'median'),
                          Mean_rating = ('vote_average', 'mean'),
                          Median_pop = ('popularity', 'median'),
                          Total_movies = ('Actors', 'count'))
```

```python
actors_data.nlargest(20, 'Total_movies')
```

```python
actors_data.nlargest(20, 'Total_revenue')
```

```python
plt.figure(figsize=(10,5))
#actors_data.Total_revenue.nlargest(10).plot(kind='bar', fontsize=15)
actors_data['Total_revenue'].sort_values(ascending=False).head(10).plot(kind='bar', fontsize=15)
plt.title('Actors', fontsize=15)
plt.ylabel('Revenue', fontsize=15)
plt.show()
```
