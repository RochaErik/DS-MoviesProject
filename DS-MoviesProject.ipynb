```python
df_best = df[['poster_path', 'title', 'budget_musd', 'revenue_musd', 'vote_count', 'vote_average', 'popularity']].copy()
```

```python
df_best
```

```python
df_best['profit_musd'] = df.revenue_musd.sub(df.budget_musd)
df_best['return'] = df.revenue_musd.div(df.budget_musd)
```

```python
df_best
```

```python
df_best.columns = ['', 'Title', 'Budget', 'Revenue', 'Votes', 'Average Rating', 'Popularity', 'Profit', 'ROI']
df_best
```

```python
df_best.set_index('Title', inplace=True)
```

```python
df_best
```

```python
df_best.sort_values(by='Average Rating', ascending=False)
```

We have movies with high average rating but with a very few number of votes

```python
df_best.sort_values(by='ROI', ascending=False)
```

As we see, the dataset contains movies with high ROI values but without any budget value. (ROI = revenue/budget)

```python
df_best.loc[df_best.Budget >= 5].sort_values(by='ROI', ascending=False)
```

```python
df_best.Budget.fillna(0, inplace=True)
df_best.Votes.fillna(0, inplace=True)
```

```python
df_best.info()
```

```python
def best_worst(n, by, ascending=False, min_budget=0, min_votes=0):
    df2 = df_best.loc[(df_best.Budget >= min_budget) & (df_best.Votes >= min_votes), 
    ['',by]].sort_values(by=by, ascending=ascending).head(n).copy()

    return HTML(df2.to_html(escape=False))
```

**Top 5 movies in revenue**

```python
best_worst(5, 'Revenue')
```

**Top 5 movies in budget**

```python
best_worst(5, 'Budget')                                                                                                 
```

**Worst 5 movies by profit**

```python
best_worst(5, 'Profit', True)
```

**Top 5 movies by average rating**

```python
best_worst(5, 'Average Rating', min_votes=20)
```

**Worst 5 movies by average rating**

```python
best_worst(5, 'Average Rating', ascending=True, min_votes=20)
```
